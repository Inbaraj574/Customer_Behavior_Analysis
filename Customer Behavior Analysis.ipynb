```python
df.columns = df.columns.str.lower()
df.coulmns = df.columns.str.replace(" ","_")
df=df.rename(columns={'purchase amount (usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Create New Columns called age_group
labels=["Young Adult","Adult","Middle-Aged","Senior"]
df["age_group"]=pd.qcut(df['age'],q=4,labels = labels)
```

```python
df[["age","age_group"]].head(10)
```

```python
#purchase_Fequency_days

frequency_mapping={
   "Fortnightly":14,
    "Weekly":7,
    "Monthly":30,
    "Quarterly":90,
    "Bi-Weekly":14,
    "Annually":365,
    "Every 3 Months":90
}
df["purchase_frequency_days"]= df["frequency of purchases"].map(frequency_mapping)
```
